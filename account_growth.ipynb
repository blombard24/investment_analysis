```python
tickers = ['qqq','voo','aapl','msft', 'atrfx', 'tqqq']


def sample(ticker, timeframe, calc_type='oldest', drip = True):
    '''
    This function outputs a dataframe with the growth of given tickers over a given 
    timeframe
    ticker=None
    timeframe=None
    calc_type="oldest"
    drip=True
    
    
    Parameters
    ----------
    
    ticker: string or list
    input a single ticker symbol or a list of ticker symbols
    
    timeframe: string to identify analysis period
    pass through value to yf.history parameter optional values include:
    1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max
    
    
    calc_type: string - calculate account value based on oldest possible date for a given 
    ticker symbol or enter "other" to calculate all starting account values on the same 
    date. e.g if a ticker symbol has a max timeframe of 1 year while others have a 
    timeframe of 5 years if calc_type = "other" all accounts will start with $10,000 
    starting on the date of the youngest ticker symbol (1 year) if calc_type = "oldest" 
    all accounts will start with $10,000 on the earliest available date
    
    drip: Dividend Reinvestment Plan - Boolean
    reinvest dividends upon distribution, share count and account value will increase
    '''
    
    #get the max date range for all tickers into data frame
    def get_date_range(ticker, timeframe, calc_type):
        df_range = pd.DataFrame()
        if calc_type == 'oldest':
            for i in ticker:
                tick_sym = yf.Ticker(i)
                temp = tick_sym.history(timeframe).index
                if len(temp) > len(df_range):
                    df_range = temp
                else:
                    pass
        else:
            for i in ticker:
                tick_sym = yf.Ticker(i)
                temp = tick_sym.history(timeframe).index
                if (len(temp) < len(df_range)) | (len(df_range) < 2) :
                    df_range = temp
                else:
                    pass
#         df_range = pd.DataFrame(df_range).set_index('Date') 
        df_range = pd.DataFrame(df_range).set_index('Date') 
        return df_range

    #Create a new df with account values for charting
    df_value = get_date_range(ticker, timeframe, calc_type)
    
    if calc_type == 'oldest':
        #iter through ticker to run calculations for each stock symbol
        for i in ticker:

            #return ticker object and price history based on ticker, and timeframe from func
            symbol = yf.Ticker(i)
            df = symbol.history(timeframe)

            #create columns for assignment later
            df[f'{i}_account_value'] = 0
            df[f'{i}_share_count'] = 0

            #assign starting value to ticker symbol
            df[f'{i}_share_count'].iloc[0] = (10000/(df['Open'].iloc[0])).copy()

            #add dividend distribution to share count if drip=True
            if drip == True:
                def DRIP():
                    df[f'{i}_share_count'] = (
                        df['Open'].add(df['Dividends']).div(df['Open'])
                        .mask(df['Dividends'].eq(0), 1).cumprod()
                        .mul(df[f'{i}_share_count'].iloc[0])
                     )
                    return df[f'{i}_share_count']
                df[f'{i}_share_count'] = DRIP()
            else:
#               get cumulative sum of share count                
                df[f'{i}_share_count'] = df[f'{i}_share_count'].cumsum()
                           
            #get account value: share_count by Open price        
            df[f'{i}_account_value'] = round((df[f'{i}_share_count'] * df['Open']),2)

            #join ticker symbol account value to date range from get_date_range() func.
            df_value = df_value.join(df[f'{i}_account_value'], how = 'left')
    else:
        #iter through ticker to run calculations for each stock symbol
        for i in ticker:
            #return ticker object and price history based on ticker, and timeframe from func
            symbol = yf.Ticker(i)
            df = symbol.history(timeframe)
            
            #Mask index based on smallest ticker index (date range)
            df = df[df.index.isin(df_value.index)].copy()

            #create columns for assignment later
            df[f'{i}_account_value'] = 0
            df[f'{i}_share_count'] = 0

            #assign starting value to ticker symbol
            df[f'{i}_share_count'].iloc[0] = (10000/(df['Open'].iloc[0])).copy()

            #add dividend distribution to share count if drip=True
            if drip == True:
                def DRIP():
                    df[f'{i}_share_count'] = (
                        df['Open'].add(df['Dividends']).div(df['Open'])
                        .mask(df['Dividends'].eq(0), 1).cumprod()
                        .mul(df[f'{i}_share_count'].iloc[0])
                     )
                    return df[f'{i}_share_count']
                df[f'{i}_share_count'] = DRIP()
            else: #if drip is false
#               get cumulative sum of share count                
                df[f'{i}_share_count'] = df[f'{i}_share_count'].cumsum()

            #get account value: share_count by Open price        
            df[f'{i}_account_value'] = round((df[f'{i}_share_count'] * df['Open']),2)

            #join ticker symbol account value to date range from get_date_range() func.
            df_value = df_value.join(df[f'{i}_account_value'], how = 'left') ###Share count needs to be deleted
        
        
    return df_value
```

```python
tickers = ['qqq','voo','aapl','msft', 'atrfx']
growth = sample(tickers,'5y',calc_type='other', drip=True)
```
